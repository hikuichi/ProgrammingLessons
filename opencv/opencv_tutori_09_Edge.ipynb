```python
#微分フィルタ系

%matplotlib inline
import cv2
import numpy as np
from matplotlib import pylab as plt

img = cv2.imread('./images/lenna.jpg', 0)

laplacian = cv2.Laplacian(img, cv2.CV_8U)#ラプラシアンフィルタ
sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=5)#ソーベルx方向、データタイプと、xの微分回数、yの微分回数、カーネルの大きさを指定
sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)#ソーベルy方向


plt.subplot(221), plt.imshow(img, 'gray')
plt.title('ORIGINAL'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(laplacian, 'gray')
plt.title('LAPLACIAN'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(sobelx, 'gray')
plt.title('SOBELX'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(sobely, 'gray')
plt.title('SOBELY'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#sobelとかlaplacianはデータタイプの指定に注意
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pylab as plt

img = cv2.imread('./images/opencv.jpg', 0)


sobel8u = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)#cv2.CV_8Uだと白から黒には反応しない
sobel64f = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)#cv2.CV_64F
abs_sobel64f = np.absolute(sobel64f)
abs_sobel8u = np.uint8(abs_sobel64f)


plt.subplot(131), plt.imshow(img, 'gray', vmin=0, vmax=255)
plt.title('ORIGINAL'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(sobel8u, 'gray', vmin=0, vmax=255)
plt.title('SOBEL8U'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(abs_sobel8u, 'gray', vmin=0, vmax=255)
plt.title('ABS SOBEL'), plt.xticks([]), plt.yticks([])
plt.show()
#なんか変な感じだな、なんでだろ
```
