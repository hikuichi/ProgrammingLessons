```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/lenna.jpg')
img2 = cv2.imread('./images/opencv.jpg')

mpltImg = cv2.merge((img[:, :, 2], img[:, :, 1], img[:, :, 0]))
mpltImg2 = cv2.merge((img2[:, :, 2], img2[:, :, 1], img2[:, :, 0]))
plt.subplot(121), plt.imshow(mpltImg, 'gray'), plt.title('LENNA')
plt.subplot(122), plt.imshow(mpltImg2, 'gray'), plt.title('LOGO')
plt.show()
```

```python
#
#ROI
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/lenna.jpg')
img2 = cv2.imread('./images/opencv.jpg')

img = img[0:200, 0:200, :]
img2 = img2[0:200, 0:200, :]

blend = cv2.addWeighted(img, 0.7, img2, 0.3, 0)
mpltImg3 = cv2.merge((blend[:, :, 2], blend[:, :, 1], blend[:, :, 0]))
plt.subplot(131), plt.imshow(mpltImg, 'gray'), plt.title('LENNA')
plt.subplot(132), plt.imshow(mpltImg2, 'gray'), plt.title('LOGO')
plt.subplot(133), plt.imshow(mpltImg3, 'gray'), plt.title('BLEND')
plt.show()
```
