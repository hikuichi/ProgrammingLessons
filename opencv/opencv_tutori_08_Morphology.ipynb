```python
#モフォロジーとか
#収縮
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白対領域となるのでINVを指定
kernel = np.ones((5, 5), np.uint8)
elosion = cv2.erode(binary, kernel, iterations = 1)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(elosion, 'gray'), plt.title('ELODED')
plt.show()
```

```python
#モフォロジーとか
#膨張
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((5, 5), np.uint8)
dilation = cv2.dilate(binary, kernel, iterations = 1)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(dilation, 'gray'), plt.title('DILATIED')
plt.show()
```

```python
#モフォロジーとか
#クロージング(膨張して、収縮)
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((9, 9), np.uint8)
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(closing, 'gray'), plt.title('CLOSING')
plt.show()
```

```python
#モフォロジーとか
#オープニング(収縮して、膨張)
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((9, 9), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(opening, 'gray'), plt.title('OPENING')
plt.show()
```

```python
#モフォロジーとか
#グラディエント(膨張と収縮の差分)
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((9, 9), np.uint8)
grad = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(grad, 'gray'), plt.title('GRADIENT')
plt.show()
```

```python
#モフォロジーとか
#トップハット(入力とオープニングの差分)
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((9, 9), np.uint8)
tophat = cv2.morphologyEx(binary, cv2.MORPH_TOPHAT, kernel)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(tophat, 'gray'), plt.title('TOPHAT')
plt.show()
```

```python
#モフォロジーとか
#ブラックハット(クロージングと入力の差分)
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./images/opencv.jpg', 0)
ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) #白が対領域となるのでINVを指定
kernel = np.ones((9, 9), np.uint8)
blackhat = cv2.morphologyEx(binary, cv2.MORPH_BLACKHAT, kernel)

plt.subplot(131), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(132), plt.imshow(binary, 'gray'), plt.title('BIN')
plt.subplot(133), plt.imshow(blackhat, 'gray'), plt.title('BLACKHAT')
plt.show()

#オープンハットとかブラックハットとかグラディエントとかいつ使うんだろう？
```
